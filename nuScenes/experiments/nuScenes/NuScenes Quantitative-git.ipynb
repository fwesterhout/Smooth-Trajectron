## Beta = 0.01

```python
model = 'int_ee_me_ped'
print(f"FDE Results for: {model}")
beta = 'beta0.01'
for ph in [2,4,6, 8]: #2, 4, got this out
    print(f"-----------------PH: {ph} -------------------")
    perf_df = pd.DataFrame()
    for f in glob.glob(f"results/int_ee_me_ped/int_ee_me_ped/beta0.01/{model}_{beta}_VEHICLE_{ph}_fde_full.csv"):
        dataset_df = pd.read_csv(f)
        dataset_df['model'] = model
        perf_df = perf_df.append(dataset_df, ignore_index=True)
    del perf_df['Unnamed: 0']

    print(f"FDE Mean @{ph*0.5}s: {perf_df['value'][perf_df['type'] == 'full'].mean()}")
    del perf_df      

    perf_df = pd.DataFrame()
    for f in glob.glob(f"results/int_ee_me_ped/int_ee_me_ped/beta0.01/{model}_{beta}_VEHICLE_{ph}_rv_full.csv"):
        dataset_df = pd.read_csv(f)
        dataset_df['model'] = model
        perf_df = perf_df.append(dataset_df, ignore_index=True)
    del perf_df['Unnamed: 0']
    print(f"RB Viols @{ph*0.5}s: {perf_df['value'][perf_df['type'] == 'full'].sum() / (len(perf_df['value'][perf_df['type'] == 'full'].index)*2000)}")
    del perf_df

    perf_df = pd.DataFrame()
    for f in glob.glob(f"results/int_ee_me_ped/int_ee_me_ped/beta0.01/{model}_{beta}_VEHICLE_{ph}_fde_most_likely_z.csv"):
        dataset_df = pd.read_csv(f)
        dataset_df['model'] = model
        perf_df = perf_df.append(dataset_df, ignore_index=True)
    del perf_df['Unnamed: 0']
    print(f"FDE @{ph*0.5}s: {perf_df['value'][perf_df['type'] == 'ml'].mean()}")      
    del perf_df

    perf_df = pd.DataFrame()
    for f in glob.glob(f"results/int_ee_me_ped/int_ee_me_ped/beta0.01/{model}_{beta}_VEHICLE_{ph}_kde_full.csv"):
        dataset_df = pd.read_csv(f)
        dataset_df['model'] = model
        perf_df = perf_df.append(dataset_df, ignore_index=True)
    del perf_df['Unnamed: 0']
    print(f"KDE @{ph*0.5}s: {perf_df['value'][perf_df['type'] == 'full'].mean()}")      
    print("----------------------------------------------")
    del perf_df
print("")
```
